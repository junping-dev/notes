## 一、仓库结构怎么管理？

假设你本地的仓库路径是：

```text
C:\Users\32840\Desktop\notes   # 示例
```

典型结构可以是这样（你可以稍微调整，但建议类似）：

```text
notes/
├── _quarto.yml        # 网站全局配置 & 导航
├── index.qmd          # 首页
├── about.qmd          # 关于（可删）
├── notebooks/         # 以后所有笔记都丢这里
│   ├── cross_entropy.ipynb
│   ├── batchnorm_layernorm.ipynb
│   └── binary_search.ipynb
├── .nojekyll
└── docs/              # quarto render 后自动生成，给 GitHub Pages 用
    ├── index.html
    └── ...
```

> **原则：**
>
> * `notebooks/` 下面随便分子目录，比如 `dl/`、`algo/`，都可以；
> * `_quarto.yml` 是“目录总控”，你每加一篇新笔记，只要在这里挂一下链接。

## 二、有了新笔记，怎么加到网站里？

分两种情况：你写的是 **Jupyter Notebook `.ipynb`** 还是 **Quarto 文档 `.qmd`**。你现在习惯是 Notebook，为你重点讲这个。

### 1. 把新的笔记文件放进仓库

例如你刚在 VSCode 里写完一个：

```text
交叉熵与分类.ipynb
```

先保存好，然后复制 / 移动到仓库里的 `notebooks/` 目录：

```text
notes/
  notebooks/
    cross_entropy.ipynb
```

（建议用英文文件名，避免以后路径问题。）

> 如果你想按主题分目录，也可以：
>
> ```text
> notebooks/
>   dl/
>     cross_entropy.ipynb
> ```

### 2. 在 Notebook 第一格写一个简单标题（可选但推荐）

在 `cross_entropy.ipynb` 的第一个 Markdown 单元里写：

```markdown
---
title: "交叉熵与分类任务"
format:
  html:
    toc: true
---

# 交叉熵与分类任务
```

前面这段 `--- ... ---` 叫 YAML 头部，Quarto 会用它当页面标题、是否显示目录等。

> **偷懒版：**
> 这一步其实可以不写，Quarto 会用文件名当标题，只是页面名字会丑一点而已。

### 3. 在 `_quarto.yml` 里把笔记挂到导航栏

打开仓库根目录的 `_quarto.yml`，大概是这样的：

```yaml
project:
  type: website
  output-dir: docs

website:
  title: "My Notes"
  navbar:
    left:
      - href: index.qmd
        text: Home

format:
  html:
    theme: cosmo
    toc: true
```

现在假设你要在导航栏加一个「Deep Learning」菜单，把这篇 notebook 放进来：

**情形 A：笔记在 `notebooks/` 下面**

```yaml
website:
  title: "My Notes"
  navbar:
    left:
      - href: index.qmd
        text: Home
      - text: Deep Learning
        menu:
          - text: 交叉熵与分类
            href: notebooks/cross_entropy.ipynb
```

**情形 B：你按子目录分，比如 `notebooks/dl/cross_entropy.ipynb`**

```yaml
website:
  title: "My Notes"
  navbar:
    left:
      - href: index.qmd
        text: Home
      - text: Deep Learning
        menu:
          - text: 交叉熵与分类
            href: notebooks/dl/cross_entropy.ipynb
```

> 注意：
>
> * `href` 写的是**相对路径**，可以直接指向 `.ipynb`；
> * `text` 是导航栏显示的中文标题。

### 4. 本地预览确认一下

在仓库目录里打开终端（或 VSCode 终端）：

```powershell
cd C:\Users\32840\Desktop\notes
quarto preview
```

浏览器会打开本地预览：

* 顶部导航栏应该能看到「Deep Learning」→「交叉熵与分类」；
* 点进去能看到你 notebook 里的代码、输出、公式。

没问题之后就可以正式发布。

## 三、每次更新后的 Git 工作流（固定 4 步）

以后你只要记住这 4 步就行：

1. **渲染网站（生成最新 docs）**

   ```powershell
   quarto render
   ```

2. **查看变更**

   ```powershell
   git status
   ```

   一般会看到：

   * 新增的 `notebooks/xxx.ipynb`
   * 更新后的 `docs/` 里的 html

3. **提交到 Git**

   ```powershell
   git add .
   git commit -m "add note: cross entropy"
   ```

4. **推送到 GitHub**

   ```powershell
   git push
   ```

GitHub Pages 会自动用最新的 `docs/` 更新，几十秒 ~ 几分钟后访问：

```text
https://junping-dev.github.io/notes/
```

就能看到新加的那篇笔记了。

## 四、给你一个“以后每加一篇笔记”的 Checklist

以后你可以照着这个小清单做（基本 5～10 分钟）：

1. 在 VSCode 里写完一个 Notebook，保存为：
   `cross_entropy.ipynb`
2. 把文件复制到仓库的 `notebooks/`（或子目录）里；
3. 在 Notebook 第一格写上标题（可选）；
4. 修改 `_quarto.yml`，在 `navbar` 里加一个链接；
5. 终端里运行：

   ```powershell
   quarto render
   git add .
   git commit -m "add note: cross entropy"
   git push
   ```

OK，这一篇就“接”到你的在线笔记站上了。
